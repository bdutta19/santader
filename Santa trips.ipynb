```python
import pandas as pd
import numpy as np

import sklearn.cluster
from shapely.geometry import MultiPoint
from geopy.distance import great_circle
```

```python
df = pd.read_csv("gifts.csv")
```

```python
df.columns
```

```python
minwg = 10
maxwg = 1000

nplat = 90
nplon = 0

def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

```python
coords = df.as_matrix(columns=['Latitude', 'Longitude'])
```

```python
kms_per_radian = 6371.0088
epsilon = 50/ kms_per_radian
epsilon
```

```python
db = sklearn.cluster.DBSCAN(eps=epsilon, min_samples=2, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))

#db = sklearn.cluster.DBSCAN(eps=epsilon, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
```

```python
cluster_labels = db.labels_
num_clusters = len(set(cluster_labels))
clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])
print('Number of clusters: {}'.format(num_clusters))
```

```python
cl_len=0 
for cl_ in clusters: 
    cl_len += len(cl_)
cl_len  
```

```python
def get_centermost_point(cluster):
    centroid = (MultiPoint(cluster).centroid.x, MultiPoint(cluster).centroid.y)
    centermost_point = min(cluster, key=lambda point: great_circle(point, centroid).m)
    return tuple(centermost_point)
```

```python
#centermost_points = clusters.map(get_centermost_point)
idx = -1
cntrmostpts = []
cluslen = 0
for clus_ in clusters:
    idx += 1
    print(idx, len(clus_))
    cluslen += len(clus_)
    if len(clus_) > 0:
        ccmpt = get_centermost_point(clus_)
        cntrmostpts.append(ccmpt)
```

```python
# Now calculate trips by partiioning upto maxwg
trips = []
clusterm = []
tripwg = 0
tripar_ = []
giftid = []

for clus_ in clusters:
    clusterm = clus_
    for cl_ in clusterm:
        dfrow = df[(df["Latitude"] == cl_[0]) & (df["Longitude"] ==  cl_[1])]
        wght = dfrow["Weight"].values[0]
        gift_ = dfrow["GiftId"].values[0]
        tripwg +=wght
        if tripwg < maxwg:
            tripar_.append(gift_)
        else:
            trips.append(set(tripar_))
            tripar_ = []
            tripwg = 0
    
```
